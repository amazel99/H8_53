```python
df.groupby("channel_name")[["publish_time", "trending_time"]].sum()
```

```python
df_smtown_topview = df_summary[
    (df_summary["channel_name"] == "SMTOWN")
    & (df_summary.like > df_summary.dislike)
]
```

```python
df_smtown_topview.sort_values(["like", "dislike"], ascending=False)
```
